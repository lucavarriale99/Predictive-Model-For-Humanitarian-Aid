### Load results and real data for comparisons

```python
import pandas as pd

results = pd.read_csv('results_all_countries.csv')
tests = pd.read_csv('ground_truth_all_countries.csv')

tests.drop('MY', axis=1, inplace=True)
results.drop('MY', axis=1, inplace=True)

# Compute the maximum value for the fatalities among all Sahel countries (visualization reasons)
maximum = max(tests.max().max(), results.max().max())
maximum
```

### Dashboard creation and visualization

```python
import matplotlib.pyplot as plt
import geopandas as gpd
import numpy as np
from ipywidgets import interactive, widgets
import ipywidgets
from IPython.display import display


world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
for index, row in results.iterrows():
    world[f'Value_{index}'] = np.nan
    for country in row.index:
        world[f'Value_{index}'].loc[world.name == country] = row[country]

for index, row in tests.iterrows():
    world[f'Test_{index}'] = np.nan
    for country in row.index:
        world[f'Test_{index}'].loc[world.name == country] = row[country]
        


# Function to update the plot based on the selected row
def update_plot(row_index):
    fig, ax = plt.subplots(ncols=2, figsize=(16,8))
    world[world.continent == 'Africa'].plot(
        figsize=(6, 6),
        edgecolor=u'gray',
        column=f'Value_{row_index}',
        legend=True,
        cmap='magma_r',
        vmax=maximum,
        ax = ax[0]
    )
    world[world.continent == 'Africa'].boundary.plot(edgecolor=u'gray', ax=ax[0])
    ax[0].set_title("Predicted Fatalities")
    
    world[world.continent == 'Africa'].plot(
        figsize=(6, 6),
        edgecolor=u'gray',
        column=f'Test_{row_index}',
        legend = True,
        cmap = 'magma_r',
        vmax = maximum,
        ax=ax[1]
    )
    world[world.continent == 'Africa'].boundary.plot(edgecolor=u'gray', ax=ax[1])
    ax[1].set_title("True Fatalities")
    
    ax[0].set_ylabel('Latitude')
    ax[0].set_xlabel('Longitude')
    ax[1].set_ylabel('Latitude')
    ax[1].set_xlabel('Longitude')

    plt.show(fig)
    
play = widgets.Play(
    value=0,
    min=0,
    max=len(results) - 1,
    step=1,
    interval=1500,
    description="Month Passed: ",
    disabled=False
)
slider = widgets.IntSlider(max=len(results) - 1)
l = ipywidgets.link((play, 'value'), (slider, 'value'))

w = interactive(update_plot, row_index=play)
box = widgets.VBox([widgets.Label("Months Passed:"),slider, w])
display(box)




```
